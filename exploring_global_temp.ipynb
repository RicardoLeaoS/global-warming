# importing

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np
import datetime

from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.seasonal import seasonal_decompose
sns.set(style="darkgrid")
```

```python
global_temp_df = pd.read_csv('data/GlobalTemperatures.csv')
```

```python
years = np.unique(global_temp_df.dt.apply(lambda x: x[:4]))
```

```python
global_temp_df['dt'] = global_temp_df['dt'].apply(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d'))
```

```python
global_temp_df['year'] = global_temp_df['dt'].apply(lambda x: x.year)
```

```python
global_temp_df.isnull().mean()
```

```python
global_temp_df.info()
```

```python
global_temp_df.describe()
```

### missing data

```python
missing_value_cols = [x for x in global_temp_df.columns if global_temp_df[x].isnull().any()]
```

```python
missing_value_cols
```

```python
for i in range(0, len(missing_value_cols)):
    col = missing_value_cols[i]
    print('' + str(col))
    global_temp_df[col] = global_temp_df[col].fillna(global_temp_df.groupby('year')[col].transform('mean'))
```

```python
global_temp_df.isnull().mean()
```

```python
global_temp_df.head()
```

```python
global_temp_df = global_temp_df.iloc[:, [0,1,2,-1]]
```

```python
global_temp_df.set_index('year', inplace=True)
```

### visualizing data

```python
global_temp_df.describe()
```

```python
global_temp_df['LandAverageTemperaturePlusUncertainty'] = global_temp_df['LandAverageTemperature'] + global_temp_df['LandAverageTemperatureUncertainty']
```

```python
global_temp_df['LandAverageTemperatureMinusUncertainty'] = global_temp_df['LandAverageTemperature'] - global_temp_df['LandAverageTemperatureUncertainty']
```

```python
data = global_temp_df.drop(['LandAverageTemperatureUncertainty', 'dt'], axis=1)
```

```python
plt.subplots(figsize=(20,10))

plt.title('Historical Global Temperature')

ax = sns.lineplot(data=data,ci=40, lw=3)

ax.set(xlabel='Year', ylabel='Land Average Temperature (C)')

plt.savefig('global_temp.pdf')

plt.show()
```

### autocorr

```python
shift_1 = global_temp_df['LandAverageTemperature'].autocorr(lag=1)
print(shift_1)
```

```python
shift_2 = global_temp_df['LandAverageTemperature'].autocorr(lag=2)
print(shift_2)
```

```python
shift_3 = global_temp_df['LandAverageTemperature'].autocorr(lag=10)
print(shift_3)
```

```python
fig, ax = plt.subplots(figsize=(20,10))

ax = plot_acf(global_temp_df['LandAverageTemperature'], ax, title='Global Temperature Autocorrelation')

plt.savefig('global_temp_auto_corr.pdf')

plt.show()
```

## decomposing time series

```python
global_temp_df.set_index('dt', inplace=True)
```

```python
s_decomp = seasonal_decompose(global_temp_df['LandAverageTemperature'], model='additive')
```

```python
plt.subplots(figsize=(20,10))
plt.title('Global Temperature Observed')
s_decomp.observed.plot()

plt.savefig('global_temp_decomposition_observed.pdf')



plt.show()
```

```python
plt.subplots(figsize=(20,10))
plt.title('Global Temperature Trend')

plt.savefig('global_temp_decomposition_trend.pdf')


s_decomp.trend.plot()
plt.show()
```

```python
plt.subplots(figsize=(20,10))
plt.title('Global Temperature Seasonality')

plt.savefig('global_temp_decomposition_season.pdf')

s_decomp.seasonal.plot()

plt.show()
```

```python
plt.subplots(figsize=(20,10))
plt.title('Global Temperature Resid')

plt.savefig('global_temp_decomposition_resid.pdf')

s_decomp.resid.plot()

plt.show()
```
